```python
data1['target_quality'] = "high"
data2['target_quality'] = "high"
```

```python
data1["target_quality"][data1["quality"]<5]="low"     
data2["target_quality"][data2["quality"]<5]="low"     
```

```python
#white wine

import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split

x = data1[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol"]]
y = data1["target_quality"]


[xTrain, xTest] = train_test_split(x, test_size=0.2, random_state=42) # la dimensione del test set è il 30% del dataset
[yTrain, yTest] = train_test_split(y, test_size=0.2, random_state=42) # la dimensione del test set è il 30% del dataset 


clf = LogisticRegression(solver='lbfgs', max_iter=1000)
clf.fit(xTrain[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol"]], yTrain)
y_hat = clf.predict(xTest[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol"]])

# Calculate the accuracy 
accuracy = accuracy_score(yTest, y_hat)
print("accuracy = {}".format(accuracy))


# Print precision, recall, f1-score
print(classification_report(yTest, y_hat))

# Compute and show the confusion matrix
confusion_matrixr = confusion_matrix(yTest, y_hat)

labels = ['Negative', 'Positive']

fig = plt.figure()
ax = fig.add_subplot(111)

#plot the heatmap
cax = ax.matshow(confusion_matrixr)

#write number on each square
for (i, j), z in np.ndenumerate(confusion_matrixr):
    ax.text(j, i, '{:0.1f}'.format(z), ha='center', va='center',
            bbox=dict(boxstyle='round', facecolor='white', edgecolor='0.3'))

plt.title('Confusion matrix of the classifier\n')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)

plt.show()
```

```python
#red wine 

import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split

x = data2[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol"]]
y = data2["target_quality"]


[xTrain, xTest] = train_test_split(x, test_size=0.2, random_state=42) # la dimensione del test set è il 30% del dataset
[yTrain, yTest] = train_test_split(y, test_size=0.2, random_state=42) # la dimensione del test set è il 30% del dataset 


clf = LogisticRegression(solver='lbfgs', max_iter=1000)
clf.fit(xTrain[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol"]], yTrain)
y_hat = clf.predict(xTest[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol"]])

# Calculate the accuracy 
accuracy = accuracy_score(yTest, y_hat)
print("accuracy = {}".format(accuracy))


# Print precision, recall, f1-score
print(classification_report(yTest, y_hat))

# Compute and show the confusion matrix
confusion_matrixw = confusion_matrix(yTest, y_hat)

labels = ['Negative', 'Positive']

fig = plt.figure()
ax = fig.add_subplot(111)

#plot the heatmap
cax = ax.matshow(confusion_matrixw)

#write number on each square
for (i, j), z in np.ndenumerate(confusion_matrixw):
    ax.text(j, i, '{:0.1f}'.format(z), ha='center', va='center',
            bbox=dict(boxstyle='round', facecolor='white', edgecolor='0.3'))

plt.title('Confusion matrix of the classifier\n')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)

plt.show()
```

## 2.
Apply Logistic Regression algorithm on the complete dataset, considering
together red and white wines, and adding a feature column named
"wine_type", whose elements can assume the values {'red', 'white'}.

```python
data1["wine_type"]= 1   #white
data2["wine_type"]= 0   #red
merged = pd.concat([data1,data2], ignore_index=True, sort=False)
#print(merged.head(50))
#train and test division
x = merged[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","wine_type"]]
y = merged["target_quality"]

[xTrain, xTest] = train_test_split(x, test_size=0.2, random_state=42) # la dimensione del test set è il 30% del dataset
[yTrain, yTest] = train_test_split(y, test_size=0.2, random_state=42) # la dimensione del test set è il 30% del dataset 


clf = LogisticRegression(solver='lbfgs', max_iter=1000)
clf.fit(xTrain[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","wine_type"]], yTrain)
y_hat = clf.predict(xTest[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","wine_type"]])

# Calculate the accuracy 
accuracy = accuracy_score(yTest, y_hat)
print("accuracy = {}".format(accuracy))


# Print precision, recall, f1-score
print(classification_report(yTest, y_hat))

# Compute and show the confusion matrix
confusion_matrixm = confusion_matrix(yTest, y_hat)

labels = ['Negative', 'Positive']

fig = plt.figure()
ax = fig.add_subplot(111)

#plot the heatmap
cax = ax.matshow(confusion_matrixm)

#write number on each square
for (i, j), z in np.ndenumerate(confusion_matrixm):
    ax.text(j, i, '{:0.1f}'.format(z), ha='center', va='center',
            bbox=dict(boxstyle='round', facecolor='white', edgecolor='0.3'))

plt.title('Confusion matrix of the classifier\n')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)

plt.show()
```

## 3.
Apply a Linear Regression algorithm, considering the quality measure learned as a continuous value. When evaluating the prediction, make an approximation on the quality value predicted by the linear regression model to retrieve a target_quality (with the same procedure asin point 1). Evaluate the target_quality prediction on the binned real_quality.

```python
# find correlations to target
corr_matrix = merged.corr().abs()
print(corr_matrix['quality'].sort_values(ascending=False))
```

```python
x = merged[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","wine_type"]]
y = merged["quality"]

[xTrain, xTest] = train_test_split(x, test_size=0.2, random_state=42) # la dimensione del test set è il 30% del dataset
[yTrain, yTest] = train_test_split(y, test_size=0.2, random_state=42) # la dimensione del test set è il 30% del dataset 


clf = LinearRegression()
clf.fit(xTrain[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","wine_type"]], yTrain)
y_hat = clf.predict(xTest[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","wine_type"]])


print("intercept = {}".format(clf.intercept_))
print("coeff = {}".format(clf.coef_))

# Performance evaluation
yTestMean = yTest.mean()
n = len(yTest)

SST = np.sum((yTest - yTestMean) ** 2) / n
print("SST = {}".format(SST))

SSE =  np.sum((yTest - y_hat) ** 2) / n
print("SEE = {}".format(SSE))

R_square = 1 - (SSE / SST)
print("R-square = {}".format(R_square))

print("The selected attributes explain {}% of the variation in quality".format(R_square))


# Convert to target_quality and calculate the accurancy
correct_predictions_count = 0
for index, value in enumerate(yTest):
    if y_hat[index]>=5 and value>=5 or y_hat[index]<5 and value<5:
        correct_predictions_count += 1

print("The selected attributes explain {}% of the variation in target_quality"
      .format((correct_predictions_count/len(y_hat))*100))
```

Linear regression works worse than logistic regression but it also allows us to understand the difference between a low quality wine and a high quality wine.
